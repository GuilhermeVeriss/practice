*Problema 4*

(a) contour plot

```python
# Aqui temos a distribuição de probabilidade p(X, Y|D,I)

x1d = y1d = np.linspace(0, 1, 1000)

x, y = np.meshgrid(x1d, y1d)

A1 = 4.82033
A2 = 4.43181
x1 = 0.5
y1 = 0.5
x2 = 0.65
y2 = 0.75
sig1 = 0.2
sig2 = 0.04

p_xy = A1 * np.exp(-((x - x1)**2 + (y - y1)**2) / (2 * sig1**2)) + \
       A2 * np.exp(-((x - x2)**2 + (y - y2)**2) / (2 * sig2**2))

```

```python
# Contour plot
plt.contour(x, y, p_xy, levels = 30)
```

(c) Marginal probabilities

```python
# p(X|D, M, I)
marginal_x = np.array([np.sum(p_y) * np.diff(y1d)[0] for p_y in p_xy.T])

# p(X|D, M, I)
marginal_y = np.array([np.sum(p_x) * np.diff(x1d)[0] for p_x in p_xy])

plt.plot(x1d, marginal_x)
plt.plot(y1d, marginal_y)
```

```python
np.sum(p_xy) * np.diff(x1d)[0]**2
# constatamos que a probabilidade p(X, Y| D, I) Não está normalizada. 
# Então as probabilidades marginais não estarão normalizadas também
# O fator de normalização, nesse caso, seria o inverso da integral da distribuição
# p(X, Y|D, I)
```

(d) Marginal and projected probability density function of X

```python


project_x = np.array([np.max(p) for p in p_xy.T])
# normalized projected X
project_x /= np.sum(project_x) * np.diff(x1d)[0] 

# normalized X marginalization 
marginal_x /= np.sum(marginal_x) * np.diff(x1d)[0]

plt.plot(x1d, marginal_x)
plt.plot(x1d, project_x)
```

(e) Marginal and projected function for Y

```python
project_y = np.array([np.max(p) for p in p_xy])
# normalization 
project_y /= np.sum(project_y) * np.diff(y1d)[0]

# normalization of marginal probability for Y
marginal_y /= np.sum(marginal_y) * np.diff(y1d)[0]

plt.plot(y1d, marginal_y)
plt.plot(y1d, project_y)
```
